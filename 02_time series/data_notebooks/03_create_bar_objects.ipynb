```python
from cleaning_functions import *
```

# Define paths

```python
file_path = '../clean_data/time_bars_5S/'
trades_files = [file_path+f for f in os.listdir(file_path) if f.endswith('.pickle')]
```

```python
symbols = [f.replace('.pickle', '') for f in os.listdir(file_path) if f.endswith('.pickle')]
```

```python
path_dict = {}

for s in symbols:
    path_dict[s] = [x for x in trades_files if s in x][0]
```

# Load all contracts and save bar objects

```python
for s in symbols:
    print('Processing {}...'.format(s))
    
    df = pd.read_pickle(path_dict[s])
    
    if df.shape[0]>0:

        config_params = {
        'trading_session': ('00:00:05', '23:00:00'),
        'second_resolution': 5,
        'index_from_df': False,
        'verbose':True}

        config = dtypes.UserInputParser(**config_params)

        clean_obj = OilClean(df, config)
        bar = clean_obj()

        # save bar object
        bar.to_pickle('../clean_data/bar_obj_5S/{}.pickle'.format(s))
```
