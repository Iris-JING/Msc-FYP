# Load and clean data

## Load data

```python
file = '../clean_data/new_first_contract_5S.pickle'
df = pd.read_pickle(file)
```

```python
# drop useless columns
df.drop(['Date', 'Symbol_end', 'Symbol_start'], axis=1, inplace=True)
```

```python
df.isnull().sum()
```

## Add columns

```python
df['dow'] = df.index.weekday
df['time'] = df.index.time
df['time'] = df['time'].apply(lambda x : x.replace(second=0))
df['year'] = df.index.year
```

## Explore overnight return

```python
first_prices = df.query('Volume>0').groupby('date').first()
last_prices = df.query('Volume>0').groupby('date').last()
```

```python
overnight_ret = (first_prices['Open'] / last_prices['Close'].shift(1)).to_frame('overnight_ret')
overnight_ret['overnight_ret'] = np.log(overnight_ret + 1e-6)

overnight_ret.dropna(inplace=True)

overnight_ret['dow'] = overnight_ret.index.weekday
```

```python
p = line_plot(overnight_ret, '.index', 'overnight_ret', ylabeller=bp_labels,fontsize=8, figsize = (8,4))
#p['ax'].set_ylim([-0.005,0.005])
```

```python
p = hist_plot(overnight_ret.abs(), 'overnight_ret', facet_x = 'dow', var_range = [0,0.003], xlabeller=bp_labels, bins=25, fontsize=8, figsize = (8,4))
```
