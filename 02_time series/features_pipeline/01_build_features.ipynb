# Define parameters

```python
RESOLUTION='5S'
VOLUME=50

feature_save_folder = '../processed_data/features/data_vol{}_bar_{}/'.format(VOLUME, RESOLUTION)
```

# Load data

```python
bar_file='../clean_data/bar_obj_volume{}_perpetual_ts.pickle'.format(VOLUME)
bar = pd.read_pickle(bar_file)
resampled = bar.df
del bar
```

## Add useful columns

```python
resampled['wVWAP'] = resampled['Volume'] * resampled['VWAP']
resampled['wVWAP2'] = resampled['Volume'] * resampled['VWAP2']

resampled['time'] = resampled.index.time
```

## Filter useful data

```python
resampled=resampled.loc[resampled.index>=datetime.datetime(2011,1,1)]
```

## Init saver obj

```python
feature_saver = DataSaver(feature_save_folder)
```

## Init feature bar obj

```python
# Create FeatureBar
safe_mode = False
verbose = False
vb = dtypes.FeatureBar(resampled, trading_groups=['date'], safe_mode=safe_mode, verbose=verbose)
```

## Init feats dict and rows to skip

```python
feats={}
rows_to_skip=0
```

# Compute features

## Fractional differentiation

```python
non_stationary_feats = ['Open', 'Close', 'Low', 'High', 'VWAP', 'VWAP2']
```

```python
fd = FractionalDifferencing(vb.df)
df_frac = fd.fracdiff(0.3, cols=non_stationary_feats, thres=1e-5)
```

```python
fd_feats = [c.replace('0.3', '') for c in df_frac.columns]
df_frac.columns = fd_feats
```

```python
vb.df=vb.df.join(df_frac)
```

```python
vb.df[fd_feats].isnull().sum()
```

```python
feats['fd_prices'] = fd_feats
rows_to_skip = max(rows_to_skip, vb.df[fd_feats].isnull().sum().max())
```

```python
p = compare_plot( vb.df.tail(2000).reset_index(), '.index', fd_feats, fontsize=7, figsize=(6,3), lw=1)
```

```python
p = variable_histogram( vb.df[fd_feats].sample(frac=0.1), fd_feats, outlier_perc=0.001, fontsize=6, figsize=(6,3))
```

## Price Ranges

```python
vb.df['MeanHLRange'] = (vb.df['High'] - vb.df['Low'])
vb.df['NormHLRange'] = vb.df['MeanHLRange'] / vb.df['VWAP']

vb.df['MeanCORange'] = (vb.df['Close'] - vb.df['Open'])
vb.df['NormCORange'] = vb.df['MeanCORange'] / vb.df['VWAP']

for hf in [25, 50, 100, 200]:
    print('Compute trend in Range with half-life {}...'.format(hf))
    # compute EWMA
    _ = vb.set_ewma(['MeanHLRange', 'NormHLRange', 'MeanCORange', 'NormCORange'], hf)
```

```python
range_feats = [c for c in vb.df if 'Range' in c]
feats['range'] = range_feats
```

## Price moving averages

```python
# Prices EWMA (Close and VWAP)
prices_columns = ['Close', 'VWAP', 'FD_VWAP']
prices_feats=[]

ewma_half_lives = [5, 10, 25, 50, 100, 200]
for hf in ewma_half_lives:
    print('Compute smoothed prices with half-life {}...'.format(hf))
    ewma_prices = vb.set_ewma(prices_columns, hf)
    prices_feats.extend([f for f in ewma_prices])

feats['prices'] = prices_feats
print(prices_feats)
```

## Fractional differentiation of moving averages

## Price std within buckets

```python
# Intra-bucket volatility: sigma_VWAP & its EWMA

# We compute both the volatility of EWMA and smooth the volatility of VWAP
sigma_feats = []
for hf in ewma_half_lives:
    print('Compute EW volatility with half-life {}...'.format(hf))
    # volatility of smoothed VWAP
    sigma_ewma = vb.set_ewma(['wVWAP', 'wVWAP2', 'VWAP2'], hf)
    vb.df['sigma_EWMA{}R_VWAP'.format(hf)] = np.sqrt(np.round(vb.df['EWMA{}R_VWAP2'.format(hf)] 
                                                              - vb.df['EWMA{}R_VWAP'.format(hf)] ** 2, 10))
    sigma_feats.append('sigma_EWMA{}R_VWAP'.format(hf))

for hf  in [25, 50, 100, 200]:
    # smoothed volatility
    print('Compute volatility with half-life {}...'.format(hf))
    sigma_ewma = vb.set_ewma(['sigma_VWAP'], hf)
    sigma_feats.extend([f for f in sigma_ewma if 'sigma_VWAP' in f])

feats['sigma'] = sigma_feats
print(sigma_feats)
```

## Crossbar Standard Deviation & its EWMA

```python
# Crossbar Standard Deviation & its EWMA

feats['cb_vola'] = []
for window in [25, 50, 100, 200]:
    for f in ['VWAP', 'Close', 'FD_VWAP']:
        print('Compute {} VOLATILITY with window {}...'.format(f, window))
        vola_vwap = vb.set_rolling_std(price_col=f, window=window)
        feats['cb_vola'].extend(vola_vwap)
print(feats['cb_vola'])

# fill nans with zeros
vb.df[feats['cb_vola']] = vb.df[feats['cb_vola']].fillna(0)
```

## Within and Cross Bucket Volatility Deviation

```python
vola_dev_feats = []
for sigma, cb_vola, hf in zip([ 'EWMA25R_sigma_VWAP', 'EWMA50R_sigma_VWAP', 'EWMA100R_sigma_VWAP','EWMA200R_sigma_VWAP'],
                           ['CrossBarVol_VWAP_25R','CrossBarVol_VWAP_50R', 'CrossBarVol_VWAP_100R', 'CrossBarVol_VWAP_200R'],
                         [25, 50, 100, 200]):
    vb.df['VolaDev_{}_to_{}'.format(sigma.split('_')[1], cb_vola.split('_')[2])] = (vb.df[sigma] - vb.df[cb_vola]
                                                                                   )
    vb.df['VolaNorm_{}_to_{}'.format(sigma.split('_')[1], cb_vola.split('_')[2])] = (vb.df[sigma] - vb.df[cb_vola]
                                                                                   ) / vb.df['EWMA{}R_VWAP'.format(hf)]
    vola_dev_feats.extend(['VolaDev_{}_to_{}'.format(sigma.split('_')[1], cb_vola.split('_')[2]),
                          'VolaNorm_{}_to_{}'.format(sigma.split('_')[1], cb_vola.split('_')[2])])
print(vola_dev_feats)
```

## Price deviation

```python
rel, vold = [], []
for hf in ewma_half_lives:
    # Relative Deviation:
    # VWAP / EWMA(Price) - 1
    
    # VWAP
    vb.df['RelDelta_EWMA{}R_VWAP'.format(hf)] = vb.df['VWAP'] / vb.df['EWMA{}R_VWAP'.format(hf)] - 1.
    rel.append('RelDelta_EWMA{}R_VWAP'.format(hf))
    
    # FD_VWAP
    vb.df['RelDelta_EWMA{}R_FD_VWAP'.format(hf)] = vb.df['FD_VWAP'] / vb.df['EWMA{}R_FD_VWAP'.format(hf)] - 1.
    rel.append('RelDelta_EWMA{}R_FD_VWAP'.format(hf))

    # Deviation wrt Volatility: how far of is current price from trend as normalised by volatility:
    # (VWAP-EWMA(Price)) / EWMA(sigma_Price) ; note: sigma_EWMA variant is used
    
    # VWAP
    vb.df['VolatilityDelta_EWMA{}R_VWAP'.format(hf)] = (vb.df['VWAP'] - vb.df['EWMA{}R_VWAP'.format(hf)]) / vb.df[
        'sigma_EWMA{}R_VWAP'.format(hf)]
    vold.append('VolatilityDelta_EWMA{}R_VWAP'.format(hf))

# select features to be used
deviation_feats = rel + vold + vola_dev_feats
feats['deviation'] = deviation_feats
```

## Differential quantitites

```python
# Differential quantities: Relative change of a quantity

differential_feats = []

for lb in [1, 5, 10, 20, 50, 100]:
    print(lb)
    log_ret = vb.set_log_return(cols=['VWAP', 'Close', 'EWMA5R_VWAP', 'High', 'Low'], open_shift=lb, fill_nans=False)
    differential_feats.extend(log_ret)

# select features to be used
vb.df[differential_feats] = vb.df[differential_feats].replace([np.inf, -np.inf], np.nan).fillna(0.0)
feats['differential'] = differential_feats
```

```python
p = variable_histogram( vb.df[differential_feats].sample(frac=0.1), differential_feats, outlier_perc=0.001, fontsize=6, figsize=(6,3))
```

## Financial indicators

```python
indicators_feats={}
```

### Sharpe

```python
indicators_feats['sharpe'] = []

for window in [10, 25, 50, 100]:
    print(window)
    sharpes = vb.set_sharpe(cols=['VWAP', 'Close'], ret_window=window)
    fd_sharpes = vb.set_sharpe(cols=['FD_VWAP'], ret_window=window)
    indicators_feats['sharpe'].extend(fd_sharpes + sharpes)
    
# fill nans with zeros
vb.df[indicators_feats['sharpe']] = vb.df[indicators_feats['sharpe']].fillna(0)
```

### MACD

```python
indicators_feats['macd'] = []
MACD_WINDOWS = [(200, 100), (200, 20), (50, 10), (20, 10), (200, 10), (20, 5)]
for w in MACD_WINDOWS:
    for f in ['VWAP', 'Close']:
        print('Compute {} MACD with windows {} and {}...'.format(f, w[0], w[1]))
        macd_feats = vb.set_macd(price=f, large_window=w[0], small_window=w[1])
        indicators_feats['macd'].extend(macd_feats)
        
# fill nans with zeros
vb.df[indicators_feats['macd']] = vb.df[indicators_feats['macd']].fillna(0)
```

### Swing index

```python
max_price_changes = [1, 5, 10, 20, 50]
indicators_feats['si'] = []

for mpc in max_price_changes:
    print('Compute SI with MPC {}...'.format(mpc))
    si_f = vb.set_swing_index(mpc)
    indicators_feats['si'].extend(si_f)
```

### Accumulated swing index

```python
indicators_feats['asi'] = []

for mpc in max_price_changes:
    print('Compute ASI with MPC {}...'.format(mpc))
    asi_f = vb.set_acc_swing_index(mpc, groups='session')
    indicators_feats['asi'].extend(asi_f)
    
    
# fill nans with zeros
vb.df[indicators_feats['si']] = vb.df[indicators_feats['si']].fillna(0)
vb.df[indicators_feats['asi']] = vb.df[indicators_feats['asi']].fillna(0)
```

### ADX

```python
indicators_feats['adx'] = []

for hl in [5, 20, 50, 100, 150, 200, 500, 1000]:
    print('Compute ADX with half life {}...'.format(hl))
    adx_cols = vb.set_adx(hl)
    indicators_feats['adx'].extend(adx_cols)
```

### All indicators

```python
l = [v for v in indicators_feats.values()]
feats['indicators'] = [item for sublist in l for item in sublist]
feats['indicators'] = list(set(feats['indicators']))
```

## Bucket duration

```python
vb.df['LOG_bucket_duration_seconds'] = log_transform(vb.df['bucket_duration_seconds'].values, offset=1)

duration_feats = [c for c in resampled.columns if ('bucket_duration_seconds' in c)]
feats['duration'] = duration_feats
```

## Calendar feats

```python
## dow, day, wom, month
vb.df['dow'] = vb.df.index.weekday
vb.df['day'] = vb.df.index.day
vb.df['wom'] = (vb.df['day'] - 1) // 7 + 1
vb.df['month'] = vb.df.index.month

## Intra-day time

# features measuring time elapsed from beginning of the day
vb.df['dt_time'] = (np.array(vb.df.index, dtype='datetime64[s]') - np.array(vb.df.index, dtype='datetime64[D]'))
vb.df['elapsed_minutes'] = (np.floor(vb.df['dt_time'].dt.total_seconds() / 60)).astype(int)
vb.df['elapsed_minutes_cont'] = (np.floor(vb.df['dt_time'].dt.total_seconds() / 60)).astype(int)
vb.df['elapsed_5_minutes'] = 5 * (np.floor(vb.df['dt_time'].dt.total_seconds() / (60 * 5))).astype(int)

feats['calendar'] = ['dow', 'day', 'wom', 'month', 'dt_time', 'elapsed_minutes', 'elapsed_minutes_cont','elapsed_5_minutes']
```

# Save features

```python
l = [v for v in feats.values()]
all_feats = [item for sublist in l for item in sublist]
all_feats = list(set(all_feats))

print(len(all_feats))
```
